```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook, push_notebook
from bokeh.models import HoverTool, CustomJS, Slider, ColumnDataSource
from bokeh.layouts import row, column, gridplot
import trackpy as tp
from scipy.ndimage.filters import gaussian_filter
from scipy.ndimage import label
from scipy.optimize import curve_fit
output_notebook()
```

```python
file_dir = 'C:\\Users\\aquiles\\Data\\run10\\tracking\\20-06-18\\run4'
filename = 'particleDiameters4.csv'
data = pd.read_csv(os.path.join(file_dir, filename), header=None)
```

```python
p1 = figure(title="Histogram particle diameter", background_fill_color="#E8DDCB")
hist, edges = np.histogram(data[1], density=True, bins=50)
bin_centres = (edges[:-1] + edges[1:])/2
def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
p0 = [0.035, 20., 10.]
coeff, var_matrix = curve_fit(gauss, bin_centres, hist, p0=p0)
x_hist = np.linspace(0, 160, 500)
hist_fit = gauss(x_hist, *coeff)

p1.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color="#036564", line_color="#033649")
p1.line(x_hist, hist_fit, line_color='#ff8821', line_width=5)
p1.xaxis.axis_label = 'Particle diameter (nm)'
p1.yaxis.axis_label = 'Frequency'
p1.xaxis.axis_label_text_font_size = '14pt'
p1.yaxis.axis_label_text_font_size = '14pt'
p1.xaxis.major_label_text_font_size = '12pt'
p1.yaxis.major_label_text_font_size = '12pt'
show(p1)
```

```python
p1 = figure(title="Histogram particle diameter", background_fill_color="#E8DDCB")
hist, edges = np.histogram(data, density=True, bins=20)
hist2, edges2 = np.histogram(data2, density=True, bins=20)
bin_centres = (edges[:-1] + edges[1:])/2
bin_centres2 = (edges2[:-1]+edges2[1:])/2
def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
p0 = [0.035, 20., 10.]
coeff, var_matrix = curve_fit(gauss, bin_centres, hist*300, p0=p0)
coeff2, var_matrix2 = curve_fit(gauss, bin_centres2, hist2*200, p0=p0)
x_hist = np.linspace(10, 60, 500)
hist_fit = gauss(x_hist, *coeff)
hist_fit2 = gauss(x_hist, *coeff2)

p1.quad(top=hist*300, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color="#036564", line_color="#033649")
p1.quad(top=hist2*200, bottom=0, left=edges2[:-1], right=edges2[1:],
        fill_color="#032364", line_color="#033649")
p1.line(x_hist, hist_fit, line_color='#ff8821', line_width=5)
p1.line(x_hist, hist_fit2, line_color='#ffaa21', line_width=5)
p1.xaxis.axis_label = 'Particle diameter (nm)'
p1.yaxis.axis_label = 'Particle Count'
p1.xaxis.axis_label_text_font_size = '18pt'
p1.yaxis.axis_label_text_font_size = '18pt'
p1.xaxis.major_label_text_font_size = '16pt'
p1.yaxis.major_label_text_font_size = '16pt'
p1.title.text_font_size = '22pt'
show(p1)
```
